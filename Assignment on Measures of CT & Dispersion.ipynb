### 1.  Find out how many males and females participated in the test.

```python
num_females = (df["gender"] == "female").sum()
num_males = (df["gender"] == "male").sum()

print("Number of females:", num_females)
print("Number of males:", num_males)
```

### 2. What do you think about the students' parental level of education?

```python
parental_education_counts = df["parental level of education"].value_counts()
```

```python
parental_education_percentages = parental_education_counts / len(df) * 100
```

```python
print("Parental Education Counts:")
print(parental_education_counts)
print("\nParental Education Percentages:")
print(parental_education_percentages)
```

### 3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
# Calculate the average score for each gender in each subject
avg_scores = df.groupby("gender").mean()[["math score", "reading score", "writing score"]]
print(avg_scores)
```

```python
# Create a bar plot to visualize the results
avg_scores.plot(kind="bar")
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
#Math : Males
#Reading: Females
#Writing :Females
```

```python
avg_scores = df.groupby("test preparation course").mean()[["math score", "reading score", "writing score"]]
print(avg_scores)
```

```python
# Create a bar plot to visualize the results
avg_scores.plot(kind="bar")
plt.title("Average Scores by Test Preparation Course Status")
plt.xlabel("Test Preparation Course Status")
plt.ylabel("Average Score")
plt.show()
```

```python
#Math : Test course completed
#Reading: Test course completed
#Writing :Test course completed

# Hence in all aspects those students who completed the test course take the lead.
```

### 4. What do you think about the scoring variation for math, reading and writing based on 
● Gender ● Test preparation course

```python
std_scores = df.groupby(["gender"]).std()
```

```python
print(std_scores)
```

```python
std_scores.plot(kind="bar")
plt.title("Standard Deviation of Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Standard Deviation")
plt.show()
```

```python
# Here there is a slightly higher variation in scores of female students in Maths, indicating more varied range of math scores than males.
```

```python
std_scores = df.groupby([ "test preparation course"]).std()
```

```python
print(std_scores)
```

```python
std_scores.plot(kind="bar")
plt.title("Standard Deviation of Scores by Test Preparation Course Status")
plt.xlabel("Test Preparation Course Status")
plt.ylabel("Standard Deviation")
plt.show()
```

```python
# It is evident that those students who didn't complete the test preparation course are having a wider spread of scores in all subjects discussed above compared with those who had completed the same.
```
